```python
import requests

bhashini_endpoint = 'https://meity-auth.ulcacontrib.org/ulca/apis/v0/model/compute'

def translate_text(source, input_text, target_language):
    headers = {
        'Content-Type': 'application/json',
    }

    data = {
        'input': [{
            "source": input_text
        }],
        'task': 'translation',
        'modelId': '63b28348ff7cd87a3f7e1029',
        'userId': None,
    }

    response = requests.post(bhashini_endpoint, json=data, headers=headers)

    if response.status_code == 200:
         res = response.json()
         print(res['output'][0]['target'])
    else:
        print(f'Translation failed. Status code: {response.status_code}')

# Example usage:
input_text = "Hello, this is nikhil. I am student of iit roorkee"
target_language = "Hindi"  # Hindi
source = 'English'
translate_text(source, input_text, target_language)
```

```python
import base64

bhashini_endpoint = 'https://dhruva-api.bhashini.gov.in/services/inference/pipeline'

def generate_audio(source, input_text):
    headers = {
        'Content-Type': 'application/json',
        'Authorization': f""
    }

    data = {
        'inputData': {
            'input': [{
            "source": input_text
            }]
        },
        'pipelineTasks': [{
            'taskType': "tts",
            'config': {
                'gender': "female",
                # 'ai4bharat/indic-tts-coqui-misc-gpu--t4'
                'serviceId': "ai4bharat/indic-tts-coqui-indo_aryan-gpu--t4",
                'language': {
                    'sourceLanguage': source,
                },
            },
        }]
    }

    response = requests.post(bhashini_endpoint, json=data, headers=headers)

    if response.status_code == 200:
         res = response.json()
         base64_audio_content = res["pipelineResponse"][0]["audio"][0]["audioContent"]
         decoded_audio_content = base64.b64decode(base64_audio_content)

         with open("output.wav", "wb") as audio_file:
            audio_file.write(decoded_audio_content)
        
         print("completed")

    else:
        print(f'Translation failed. Status code: {response.status_code}')

# Example usage:
source = 'mr'
input_text = 'मी भारतातील सर्वात श्रीमंत माणूस आहे'
generate_audio(source, input_text)
```
